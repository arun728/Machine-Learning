```python
def run_model(model, X_train, y_train, X_test, y_test):
    
    # Fit Model Training
    log_y_train = np.log1p(y_train)
    model.fit(X_train, log_y_train)
    
    # Get Metrics
    preds = model.predict(X_test)
    preds = abs(preds)

    rmsle = np.sqrt(mean_squared_log_error(y_test, np.expm1(preds)))
    print(f'RMSLE: {rmsle}')
```

```python
def get_stacking():
    level0 = [('rf', RandomForestRegressor(n_estimators=600, max_features=600,n_jobs=-1)),
               ('catboost', CatBoostRegressor(silent=True,learning_rate=0.18, n_estimators=2000, depth=6)), 
             ('xbg',xgb.XGBRegressor(n_estimators=1000, max_depth=12, learning_rate=0.05, colsample_bytree=0.45)), 
             ('lgb',lgb.LGBMRegressor(n_estimators=10000, num_leaves=70, max_depth=20,min_child_samples=10, 
                          learning_rate=0.01, colsample_bytree=0.8,subsample=0.5, reg_alpha=0, reg_lambda=20))]
    
    level1 = LinearRegression()
    model = StackingRegressor(estimators=level0, final_estimator=level1, cv=5)
    return model

model = get_stacking()
run_model(model, X_train, y_train, X_test, y_test)
```

```python
def stacking_reg():
    level0 = [('rf',RandomForestRegressor(n_estimators=500, max_features=500)), 
             ('xbg',xgb.XGBRegressor(n_estimators=1000, max_depth=12, learning_rate=0.05, colsample_bytree=0.45)), 
             ('lgb',lgb.LGBMRegressor(n_estimators=10000, num_leaves=70, max_depth=20,min_child_samples=10, 
                          learning_rate=0.01, colsample_bytree=0.8,subsample=0.5, reg_alpha=0, reg_lambda=20)), 
             ('et',ExtraTreesRegressor(n_estimators=500, bootstrap=True, n_jobs=-1))]
    
    level1 = Ridge(4)
    model = StackingRegressor(estimators=level0, final_estimator=level1, cv=5)
    return model

model = stacking_reg()
run_model(model, X_train, y_train, X_test, y_test)
```
