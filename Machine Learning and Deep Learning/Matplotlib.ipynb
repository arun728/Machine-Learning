```python
# OOP Figure Method
```

```python
a = np.linspace(0,10,11)
```

```python
b = a**4
```

```python
a
```

```python
b
```

```python
x = np.arange(0,10)
```

```python
y = 2*x
```

```python
x
```

```python
y
```

```python
fig = plt.figure()

# Large Axes

axes1 = fig.add_axes([0,0,1,1])

axes1.plot(a,b)

axes1.set_xlim(0,8)
axes1.set_ylim(0,8000)
axes1.set_xlabel('A')
axes1.set_ylabel('B')
axes1.set_title('Power of 4')

# Small Axes

axes2 = fig.add_axes([0.2,0.5,0.25,0.25])
axes2.plot(a,b)

axes2.set_xlim(1,2)
axes2.set_ylim(0,50)
axes2.set_xlabel('A')
axes2.set_ylabel('B')
axes2.set_title('Zoomed In')

plt.show()
```

```python
# Figure Parameters

fig = plt.figure(dpi=100,figsize=(2,2))
axes1 = fig.add_axes([0,0,1,1])
axes1.plot(a,b)

fig.savefig('new_figure.png', bbox_inches='tight')
```

```python
# Subplots
```

```python
# 1D figure

fig,ax = plt.subplots(nrows=1,ncols=1)

ax.plot(x,y)
```

```python
# 2d figure 

fig,axes = plt.subplots(nrows=2,ncols=2)
axes[0][0].plot(x,y)
axes[1][1].plot(a,b)
```

```python
# 1d figure with 3 rows

fig,axes = plt.subplots(nrows=3,ncols=1)
for ax in axes:
    ax.plot(x,y)
```

```python
# 1d figure with 3 rows

# Width and Height adjustment of Subplots

fig,axes = plt.subplots(nrows=2,ncols=2,figsize=(4,4),dpi=80)

axes[0][0].plot(x,y)

axes[0][1].plot(x,y)

axes[1][0].plot(x,y)
axes[1][0].set_ylabel('Y Label 1,0')

axes[1][1].plot(x,y)
axes[1][1].set_title('Title 1,1')


fig.suptitle('MAIN TITLE',fontsize=20)


plt.tight_layout()

fig.set_figwidth(10)

fig.subplots_adjust(top=0.8,wspace=0.5,hspace=0.5)

fig.savefig('new_subplots.png',bbox='tight')
```

```python
# Legends
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,label='X vs X')

ax.plot(x,x**2, label='X vs X^2')

ax.legend(loc=(1,0.5))
```

```python
# Colors and Line Styles
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='cyan',label='X vx X')

ax.plot(x,x+1,color='purple',label='X vs X+1')

ax.legend(loc=(1,0.5))
```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

ax.plot(x,x,color='red',lw=2,ls='--')

ax.plot(x,x+1,color='blue',linewidth=2,linestyle='-.')

```

```python
fig = plt.figure()

ax = fig.add_axes([0,0,1,1])

#ax.plot(x,x,color='red',lw=2,ls='--')

lines = ax.plot(x,x+1,color='blue',linewidth=5)
lines[0].set_dashes([1,2,1,2,10,2])
```
