```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
# Scatter Plot
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2', size='salary')
```

```python
plt.figure(figsize=(10,4),dpi=200)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2', s=200, alpha=0.4)
```

```python
plt.figure(figsize=(10,4),dpi=100)
sns.scatterplot(x='salary',y='sales',data=df,hue='level of education',palette='Dark2', s=200, style='level of education')
plt.savefig('scatterplot.png')
```

```python
np.random.seed(42)
ages = np.random.randint(0,100,200)
```

```python
sns.displot(data=ages,x='age',rug=True,bins=30,kde=True)
```
