# Splitting the Product Names and Product Keys into Separate Lists

```python
tds = soup.find_all('td')
```

```python
tds
```

```python
soup.select('.confluenceTd')[2].getText()
```

```python
rows = []

table = soup.find("table")

table_rows = table.find_all('tr')

for tr in table_rows:
    td = tr.find_all('td')
    for item in td:
        row = item.text
        rows.append(row)
        
```

```python
rows
```

```python
product_names = []
product_keys = []

for row in rows:
    if 'prod' in row:
        product_keys.append(row)
    elif ' ' in row and '/' not in row and 'In Dev' not in row:
        product_names.append(row)
    
    
```
