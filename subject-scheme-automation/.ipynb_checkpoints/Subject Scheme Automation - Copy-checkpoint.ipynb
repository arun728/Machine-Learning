```python
rows = []

table = soup.find("table")

table_rows = table.find_all('tr')

for tr in table_rows:
    td = tr.find_all('td')
    for item in td:
        row = item.text
        rows.append(row)
        
```

```python
rows
```

```python
product_names = []
product_keys = []

for row in rows:
    if 'prod-' in row:
        product_keys.append(row)
    elif ' ' in row and '/' not in row and 'In Dev' not in row or 'Agent' in row:
        product_names.append(row)
    
    
```

```python
product_names = []
product_keys = []

table = soup.find("table")

table_rows = table.find_all('tr')

for tr in table_rows:
    td = tr.find_all('td')
    row = [item.text for item in td]
    col1 = row[0:1]
    col2 = row[1:2]
    
    col1 = ''.join(col1)
    col2 = ''.join(col2)
    
    product_names.append(col1)
    product_keys.append(col2)
```
